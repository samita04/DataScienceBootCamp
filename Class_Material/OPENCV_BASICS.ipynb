```python
import cv2
img = cv2.imread('APPLE .jpeg')
start_point=(100,100)
end_point=(1200,900)
#format is BGR
color=(0,0,255)
thickness=2

image=cv2.line(img,start_point,end_point,color,thickness)
cv2_imshow(image)
```

```python
#draw a circle
import cv2
img = cv2.imread('APPLE .jpeg')
center_coordinates = (600, 450)
radius = 100
end_point=(1200,900)
#format is BGR
color=(0,0,255)
thickness=2

image=cv2.circle(img,center_coordinates, radius,color,thickness)
cv2_imshow(image)
```

```python
#draw a square
import cv2
img = cv2.imread('APPLE .jpeg')
start_point=(1,1)
end_point=(900,900)
#format is BGR
color=(0,0,255)
thickness=2

image=cv2.rectangle(img,start_point,end_point,color,thickness)
cv2_imshow(image)
```

```python
#write a text on image
import cv2
img = cv2.imread('APPLE .jpeg')
font = cv2.FONT_HERSHEY_SIMPLEX
org = (50, 250)
fontScale = 1

#format is BGR
color=(0,0,255)
thickness=2

image=cv2.putText(img, 'This is apple tree.', org, font, fontScale, color, thickness, cv2.LINE_AA)
cv2_imshow(image)
```
